```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import make_pipeline
from category_encoders import OneHotEncoder
from sklearn.metrics import f1_score, precision_score, recall_score

from sklearn.linear_model import LogisticRegression
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier, RandomForestClassifier, BaggingClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import RFE

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

### Feature Engineer

```python
# combining all features that use 'secondary_use'

earthquake['secondary_use'] = earthquake['has_secondary_use_agriculture'] + earthquake['has_secondary_use_gov_office'] + earthquake['has_secondary_use_health_post'] + \
                              earthquake['has_secondary_use_hotel'] + earthquake['has_secondary_use_industry'] + earthquake['has_secondary_use_institution'] + \
                              earthquake['has_secondary_use_other'] + earthquake['has_secondary_use_rental'] + earthquake['has_secondary_use_school'] + earthquake['has_secondary_use_use_police']
```

```python
# combining different home type features

earthquake['stone_structure'] = earthquake['has_superstructure_cement_mortar_stone'] + earthquake['has_superstructure_mud_mortar_stone'] + earthquake['has_superstructure_stone_flag']
earthquake['brick_structure'] = earthquake['has_superstructure_cement_mortar_brick'] + earthquake['has_superstructure_mud_mortar_brick']
earthquake['rc_structure'] = earthquake['has_superstructure_rc_engineered'] + earthquake['has_superstructure_rc_non_engineered']
earthquake['cement_mortar_structure'] = earthquake['has_superstructure_cement_mortar_brick'] + earthquake['has_superstructure_cement_mortar_stone']
earthquake['mud_mortar_structure'] = earthquake['has_superstructure_mud_mortar_brick'] + earthquake['has_superstructure_mud_mortar_stone'] + earthquake['has_superstructure_adobe_mud']
```

```python
# see features mostly correlated with 'damage_grade' w/ engineered features

plt.figure(figsize=(8,12))
sns.heatmap(earthquake.corr()[['damage_grade']].sort_values(by = 'damage_grade', ascending = False),annot=True )
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(earthquake.corr(), annot = True)
```

### Baseline Model

```python
# baseline model

train_labels['damage_grade'].value_counts(normalize = True)
```

```python
# # establish X + y

# X = earthquake[['mud_mortar_structure', 'stone_structure', 'count_floors_pre_eq', 'count_families', 'age', 'height_percentage', 'area_percentage', 'foundation_type', 
#                 'ground_floor_type', 'position', 'secondary_use', 'land_surface_condition']]
# y = earthquake['damage_grade']
```

```python
# establish X + y

X = earthquake.drop(columns = ['damage_grade', 'building_id'])
y = earthquake['damage_grade']
```

```python
# tts

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 123)
```

# Modeling w/ OHE

```python
# Random Forest
# https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html
# https://git.generalassemb.ly/DSI-322/6.04_Forests_and_Features/blob/master/solution-code.ipynb
# https://git.generalassemb.ly/vleong1/Submissions/blob/master/Projects/project-2-master/4.%20Neighborhood%20Modeling.ipynb

pipe_forest = make_pipeline(OneHotEncoder(use_cat_names = True), StandardScaler(), RandomForestClassifier(n_jobs = -1))

params = {'randomforestclassifier__max_depth' : [2, 3, 4, 5, 6]}

grid_forest = GridSearchCV(pipe_forest, param_grid = params)

grid_forest.fit(X_train, y_train)

print(f'Train Score: {grid_forest.score(X_train, y_train)}')
print(f'Test Score: {grid_forest.score(X_test, y_test)}')

grid_forest.best_params_
```

```python
forest_pred = grid_forest.predict(X_test)
f1_score(y_test, forest_pred, average = 'micro')
```

```python
precision_score(y_test, forest_pred, average = 'micro')
```

```python
recall_score(y_test, forest_pred, average = 'micro')
```

```python
# Extra Trees
# referenced https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.ExtraTreesClassifier.html

pipe_trees = make_pipeline(OneHotEncoder(use_cat_names = True), StandardScaler(), ExtraTreesClassifier(n_jobs = -1))

params = {'extratreesclassifier__max_depth' : [2, 3, 4, 5, 6, 7]}
grid_trees = GridSearchCV(pipe_trees, param_grid = params)

grid_trees.fit(X_train, y_train)

print(f'Train Score: {grid_trees.score(X_train, y_train)}')
print(f'Test Score: {grid_trees.score(X_test, y_test)}')

grid_trees.best_params_
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html
# average = 'micro' -- Calculate metrics globally by counting the total true positives, false negatives and false positives

trees_pred = grid_trees.predict(X_test)
f1_score(y_test, trees_pred, average = 'micro')
```

## Trying OHE with Imbalanced Data

```python
# https://git.generalassemb.ly/DSI-322/5.01-lesson-imbalanced-data/blob/master/imbalanced-data-post-class.ipynb

from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
from imblearn.pipeline import make_pipeline  # https://stackoverflow.com/questions/58825053/smote-function-not-working-in-make-pipeline
```

```python
# random forest w/ SMOTE

pipe_forest3 = make_pipeline(OneHotEncoder(use_cat_names = True), StandardScaler(), RandomOverSampler(), RandomForestClassifier(n_jobs = -1))

params = {'randomforestclassifier__max_depth' : [4, 5, 6, 7, 8, 9]}

grid_forest3 = GridSearchCV(pipe_forest3, param_grid = params)

grid_forest3.fit(X_train, y_train)

print(f'Train Score: {grid_forest3.score(X_train, y_train)}')
print(f'Test Score: {grid_forest3.score(X_test, y_test)}')

grid_forest3.best_params_
```

```python
# Extra Trees
# referenced https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.ExtraTreesClassifier.html

pipe_trees3 = make_pipeline(OneHotEncoder(use_cat_names = True), StandardScaler(), RandomOverSampler(), ExtraTreesClassifier(n_jobs = -1))

params = {'extratreesclassifier__max_depth' : [4, 5, 6, 7, 8, 9]}

grid_trees3 = GridSearchCV(pipe_trees3, param_grid = params)

grid_trees3.fit(X_train, y_train)

print(f'Train Score: {grid_trees3.score(X_train, y_train)}')
print(f'Test Score: {grid_trees3.score(X_test, y_test)}')

grid_trees3.best_params_
```

```python
# Logistic Regresion

pipe_lgr = make_pipeline(OneHotEncoder(use_cat_names = True), StandardScaler(), LogisticRegression(n_jobs = -1))

params = {'logisticregression__C' : [0.1, 0.5, 0.75, 1, 10]}

grid_lgr = GridSearchCV(pipe_lgr, param_grid = params)

grid_lgr.fit(X_train, y_train)

print(f'Train Score: {grid_lgr.score(X_train, y_train)}')
print(f'Test Score: {grid_lgr.score(X_test, y_test)}')

grid_lgr.best_params_
```

```python
X2 = train_enc.drop(columns = 'building_id')
y2 = train_labels['damage_grade']
```

```python
# tts

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, stratify = y2, random_state = 123)
```

```python
# Random Forest -- BEST MODEL
# https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html
# https://git.generalassemb.ly/DSI-322/6.04_Forests_and_Features/blob/master/solution-code.ipynb
# https://git.generalassemb.ly/vleong1/Submissions/blob/master/Projects/project-2-master/4.%20Neighborhood%20Modeling.ipynb

pipe_forest2 = make_pipeline(StandardScaler(), RandomForestClassifier(n_jobs = -1))

params = {'randomforestclassifier__max_depth' : [6, 7, 8, 9, 10, 11],
          'randomforestclassifier__max_features' : [15, 20, 30, 35]}

grid_forest2 = GridSearchCV(pipe_forest2, param_grid = params)

grid_forest2.fit(X_train2, y_train2)

print(f'Train Score: {grid_forest2.score(X_train2, y_train2)}')
print(f'Test Score: {grid_forest2.score(X_test2, y_test2)}')

grid_forest2.best_params_
```

```python
forest_pred2 = grid_forest2.predict(X_test2)
f1_score(y_test2, forest_pred2, average = 'micro')
```

```python
precision_score(y_test2, forest_pred2, average = 'micro')
```

```python
recall_score(y_test2, forest_pred2, average = 'micro')
```

```python
# Extra Trees
# referenced https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.ExtraTreesClassifier.html

pipe_trees2 = make_pipeline(StandardScaler(), ExtraTreesClassifier(n_jobs = -1))

params = {'extratreesclassifier__max_depth' : [2, 3, 4, 5, 6, 7]}

grid_trees2 = GridSearchCV(pipe_trees2, param_grid = params)

grid_trees2.fit(X_train2, y_train2)

print(f'Train Score: {grid_trees2.score(X_train2, y_train2)}')
print(f'Test Score: {grid_trees2.score(X_test2, y_test2)}')

grid_trees2.best_params_
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html
# average = 'micro' -- Calculate metrics globally by counting the total true positives, false negatives and false positives

trees_pred2 = grid_trees2.predict(X_test2)
f1_score(y_test2, trees_pred2, average = 'micro')
```
