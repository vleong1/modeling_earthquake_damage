```python
# read in data
test_values = pd.read_csv('data/Proj5_test_values.csv')
train_values = pd.read_csv('data/Proj5_train_values.csv')
train_labels = pd.read_csv('data/Proj5_train_labels.csv')
```

#### Baseline + TTS

```python
# establish X + y

X = train_values_enc
y = train_labels['damage_grade']
```

```python
# tts

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 123)
```

#### Modeling

```python
# # COMPETITION MODEL

forest_no_scale = RandomForestClassifier(n_jobs = -1, random_state = 123, max_depth = 11, max_features = 35)
forest_no_scale.fit(X_train, y_train)
```

```python
forest_no_scale.score(X_test, y_test)
```

```python
preds = forest_no_scale.predict(X_test)
```
